## Q1. Write a python program to return the buying and selling price for a stock from list of prices. 
The input should be a list of prices and the output should be the buying and selling price of the stock. 
You should buy the stock at cheap price and sell the stocks when the price is high. 

```python
prices = [59,80,90,98]
buying_price = min(prices)
print(buying_price)
selling_price = max(prices)
print(selling_price)
```

## Q2. Write python a function that checks whether all numbers in a list are greater than 10 using inbuilt functions.

```python
lst=[5,7,10,11,15]
def greater():
    result=list(filter(lambda x:x>10,lst))
    print(result)
greater()

```
