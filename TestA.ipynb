### 8.using map and lambda,convert a list so negative numbers become positive,zero becomes 'Zero',and positive numbers remain unchanged. 

```python
lst=[-2,-4,6,9,0]
result=list(map(lambda x:-x,lst))
print(result)
result2=list(map(lambda x:x>0,result))
print(result2)
result3=list(map(lambda x:'Zero' if x==0 else x,result))
print(result3)
```

### 9.use filter and lambda to keep only values that are multiple of 3 and also leave a remainder of 1 when divided by 4.

```python
lst=[2,3,4,5,6,7,8]
result=list(filter(lambda x:x%3,lst))
print(result)
result2=list(map(lambda x:1 if x%4 else x,lst))
print(result2)
```

### 11.use reduce to return the difference between the sum of odd numbers and sum of even numbers.

```python
from functools import reduce
lst=[2,4,5,6,7]
result=list(filter(lambda x:x%2==0,lst))
print(result)
r1=reduce(lambda x,y:x+y,result)
print(r1)
result2=list(filter(lambda x:x%2!=0,lst))
print(result2)
r2=reduce(lambda x,y:x+y,result2)
print(r2)
final_result=r1-r2
print(final_result)
```
